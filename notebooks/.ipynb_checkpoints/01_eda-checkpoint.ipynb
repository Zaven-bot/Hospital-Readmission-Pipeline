```python
# 1. Imports & Config
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
# 2. Load Data
df = pd.read_csv("../data/raw/hospital_readmissions.csv")
print("Shape of dataset:", df.shape)

df.head()
```

```python
# 3. Replace '?' with NaN for easier handling
df.replace("?", np.nan, inplace=True)
```

```python
# 4. Quick Summary
df.info()
df.describe(include='all').T
```

```python
# 5. Target Column Exploration
df['readmitted'].value_counts()
df['readmitted_binary'] = df['readmitted'].apply(lambda x: 1 if x == '<30' else 0)
df['readmitted_binary'].value_counts(normalize=True)
```

```python
# 6. Visualize Class Imbalance
sns.countplot(x='readmitted_binary', data=df)
plt.title("Readmission (<30 days) Distribution")
plt.xticks([0, 1], ['Not Readmitted <30', 'Readmitted <30'])
plt.show()
```

```python
# 7. Check Missing Data
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

```python
# 8. Drop Candidate Columns  Let's inspect them
columns_to_inspect = ['weight', 'payer_code', 'medical_specialty', 'examide', 'citoglipton']
for col in columns_to_inspect:
    print(f"\n--- {col} ---")
    print(df[col].value_counts(dropna=False))
```

```python
# 9. Check Unknown Categories in Gender
df['gender'].value_counts(dropna=False)
```

```python
# 10. Categorical Variables  Preview
cat_cols = ['race', 'age', 'admission_type_id', 'discharge_disposition_id', 'admission_source_id']
for col in cat_cols:
    print(f"\n{col}:\n", df[col].value_counts(dropna=False))
```

```python
# 11. Lab, Procedure, and Medication Counts
numeric_cols = ['num_lab_procedures', 'num_procedures', 'num_medications', 'number_inpatient', 
                'number_outpatient', 'number_emergency', 'time_in_hospital', 'number_diagnoses']
df[numeric_cols].hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# 12. Diagnosis Code Distribution (diag_1, diag_2)
for col in ['diag_1', 'diag_2']:
    print(f"\n{col} Top 10 values:")
    print(df[col].value_counts().head(10))
    print("Missing:", df[col].isna().sum())
```

```python
# 13. Diabetes Medications + Changes
med_cols = ['insulin', 'metformin', 'change', 'diabetesMed']
for col in med_cols:
    print(f"\n{col}:")
    print(df[col].value_counts(dropna=False))
```
